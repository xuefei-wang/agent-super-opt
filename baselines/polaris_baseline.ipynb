```python
import os
import sys
import json

project_root = os.path.abspath(os.path.join(os.getcwd(), "..")) # scratch folder
if project_root not in sys.path:
    sys.path.append(project_root)


import numpy as np
from src.spot_detection import DeepcellSpotsDetector
from src.data_io import ImageData

from dotenv import load_dotenv

load_dotenv()
```

```python
load_dotenv()

spots_val = np.load('../data/SpotNet-v1_1/val.npz', allow_pickle=True)
spots_test = np.load('../data/SpotNet-v1_1/test.npz', allow_pickle=True)


## Validation
val_images = ImageData(raw = spots_val['X'], batch_size = spots_val['X'].shape[0], image_ids = [i for i in range(spots_val['X'].shape[0])])
val_spots_truth = spots_val['y']

def preprocess_images(images: ImageData) -> ImageData:
    return images

val_images = preprocess_images(val_images)

deepcell_spot_detector = DeepcellSpotsDetector()

# Predict spots
val_pred = deepcell_spot_detector.predict(val_images)

# Get metrics
val_metrics = deepcell_spot_detector.evaluate(val_pred, val_spots_truth)


## Test
test_images = ImageData(raw = spots_test['X'], batch_size = spots_test['X'].shape[0], image_ids = [i for i in range(spots_test['X'].shape[0])])
test_spots_truth = spots_test['y']

def preprocess_images(images: ImageData) -> ImageData:
    return images

test_images = preprocess_images(test_images)

deepcell_spot_detector = DeepcellSpotsDetector()

# Predict spots
test_pred = deepcell_spot_detector.predict(test_images)

# Get metrics
test_metrics = deepcell_spot_detector.evaluate(test_pred, test_spots_truth)

metrics_filepath = "expert_baseline_metrics.json"

with open(metrics_filepath, "w") as f:
    json_output = {
        "expert_baseline_val_f1": val_metrics['f1_score'],
        "expert_baseline_test_f1": test_metrics['f1_score'],
    }
    json.dump(json_output, f)
```
